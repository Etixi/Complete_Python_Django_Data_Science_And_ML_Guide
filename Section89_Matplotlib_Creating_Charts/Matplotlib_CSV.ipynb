```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.style.use('ggplot')
```

```python
df = pd.read_csv('jersey-rainfall-1984-to-2021.csv', index_col=0)
df.head()
```

```python
df.plot(y='YearTotalmm', figsize=(15, 12), color='blue')
plt.show()
```

```python
df.plot(y='Apr', figsize=(15, 12), color='blue')
plt.show()
```

```python
df.plot(y=['Apr', 'May'], figsize=(15, 12), kind='area')
plt.show()
```

```python
df.drop(columns=['YearTotalmm']).plot(figsize=(15,12))
plt.show()
```

```python
df.T.head()
```

```python
df.drop(columns=['YearTotalmm']).T.plot(y=[1901, 1902], kind='area')
plt.show()
```

```python
df.drop(columns=['YearTotalmm']).T.plot(y=[1901, 1902], kind='bar')
plt.show()
```

```python
df.drop(columns=['YearTotalmm']).T.plot(y=[1901, 1902], kind='barh')
plt.show()
```
