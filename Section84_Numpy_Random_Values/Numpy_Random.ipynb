# **Generating Random Elements Using randint and uniform**

```python
np.random.randint(10, 20, size=(4, 5))
```

```python
np.random.uniform(10.0, 50.0, size=(4, 5))
```

# **Understaning Seed Number**

```python
np.random.seed(10)
np.random.random((3, 4, 10, 5))
```

```python
np.random.seed(1)
np.random.randint(10, 20, size=(4, 5))
```

```python
# Seed value is generated automatically
np.random.uniform(10.0, 50.0, size=(4, 5))
```
